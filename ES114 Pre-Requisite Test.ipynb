3.Fibonacci Sequence:
Write a Python function that takes a number n and returns a list of the first n Fibonacci numbers.
Plot the Fibonacci sequence using a line graph.

```python
#Question 3
def fib(n):
    l = []
    l.append(0)
    l.append(1)
    for i in range(2,n):
        l.append(l[i-1] + l[i-2])
    return l
n = int(input())
l = fib(n)
l2 =[]
for i in range(1,n+1):
    l2.append(i)
l2 = np.array(l2)

plt.plot(l2,l)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Fibbonacci Sequence Line Graph")
plt.show()
```

13.Sorting Dictionaries:
Write a function to sort a list of dictionaries by a key.

Example:

data = [{'name': 'Alice', 'age': 25}, {'name': 'Bob', 'age': 22}]
Sort by age → [{'name': 'Bob', 'age': 22}, {'name': 'Alice', 'age': 25}]

```python
#Question 13

def list_dict_sort_by_age(data):
    return sorted(data, key=lambda d: d['age'])

data = [{'name': 'Alice', 'age': 25}, {'name': 'Bob', 'age': 22}]

print(list_dict_sort_by_age(data))
```

22.Normalization:
Write a Python function to normalize a list of numbers to a range of 0 to 1.
Visualize the normalized data using a line plot.

```python
#Question 22

def normalized_plot(data):
    m = min(data)
    M = max(data)
    
    l = [(x - m) / (M - m) for x in data]
    
    plt.figure(figsize=(10, 6))
    plt.plot(l, marker='o', linestyle='-', color='b')
    plt.title('Normalized Data (Range: 0 to 1)')
    plt.xlabel('Index')
    plt.ylabel('Normalized Value')
    plt.grid(True)
    plt.show()
    return l
normalized_plot([69,100,420,1,2,3,4,5,6,10,11,12,13,14,15,15,6,161,616])
```
