1.List Operations and Basic Statistics:
Create a list of the first 20 natural numbers.
Compute the sum of all even numbers in the list.
Compute the product of all odd numbers in the list.
Calculate the mean, median, and standard deviation of the list.

```python
#Question 1

l = []
s = 0
p = 1
for i in range(1,21):
    l.append(i)
    if i%2==0:
        s+= i
    else:
        p*=i
mean = st.mean(l)
median = st.median(l)
std_dev = st.pstdev(l)
print(l,"\n",s,p,mean,median,std_dev)
```

3.Fibonacci Sequence:
Write a Python function that takes a number n and returns a list of the first n Fibonacci numbers.
Plot the Fibonacci sequence using a line graph.

```python
#Question 3
def fib(n):
    l = []
    l.append(0)
    l.append(1)
    for i in range(2,n):
        l.append(l[i-1] + l[i-2])
    return l
n = int(input())
l = fib(n)
l2 =[]
for i in range(1,n+1):
    l2.append(i)
l2 = np.array(l2)

plt.plot(l2,l)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Fibbonacci Sequence Line Graph")
plt.show()
```

6.Palindrome Check:
Write a program to check if a string is a palindrome (ignoring spaces and case).
Input: "A man a plan a canal Panama"
Output: True

```python
#Question 6

x = (input().replace(' ','')).lower()
y = True
n = len(x)
for i in range((n+1)//2):
    if(x[i] != x[n-i-1]):
        y = False
        break
print(y)
```

22.Normalization:
Write a Python function to normalize a list of numbers to a range of 0 to 1.
Visualize the normalized data using a line plot.

```python
#Question 22

def normalized_plot(data):
    m = min(data)
    M = max(data)
    
    l = [(x - m) / (M - m) for x in data]
    
    plt.figure(figsize=(10, 6))
    plt.plot(l, marker='o', linestyle='-', color='b')
    plt.title('Normalized Data (Range: 0 to 1)')
    plt.xlabel('Index')
    plt.ylabel('Normalized Value')
    plt.grid(True)
    plt.show()
    return l
normalized_plot([69,100,420,1,2,3,4,5,6,10,11,12,13,14,15,15,6,161,616])
```
